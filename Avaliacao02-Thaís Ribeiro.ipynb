```python
# 3) Execução dos algoritmos:
x = pd.get_dummies(df[['englishSpeaker','courseInstructor','course','summerOrRegularSemester','classSize']])
y = df['classAttribute']

from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.25, random_state=0)

# a. Listagem dos algoritmos aplicáveis;

import numpy as np
#Ao rodar sem gerar os dummies o score estava em 0.34 apos aumentou para 0.5

# Naïve Bayes
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB(fit_prior=True)
nb.fit(x_treino, y_treino)
nb.score(x_teste, y_teste)
print(nb.score(x_teste, y_teste))

# Decision Tree
from sklearn.tree import DecisionTreeClassifier
tr = DecisionTreeClassifier()
tr.fit(x_treino, y_treino)
tr.score(x_teste, y_teste)
print(tr.score(x_teste, y_teste))

#AdaBoost
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
ada.fit(x_treino, y_treino)
ada.score(x_teste, y_teste)
print(ada.score(x_teste, y_teste))

#OneVsRest
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
ovr = OneVsRestClassifier(LinearSVC(random_state = 2, max_iter=300000))
ovr.fit(x_treino, y_treino)
ovr.score(x_teste, y_teste)
print(ovr.score(x_teste, y_teste))

# b. Escolha pelo menos um algoritmo ainda não visto na disciplina e justifique o

```

```python
# d. Uso de k-folding;
from sklearn.model_selection import cross_val_score
from scipy.stats import bayes_mvs
import numpy as np

k = 7

scores_nb = cross_val_score(nb, x_teste, y_teste, cv = k)
scores_tr = cross_val_score(tr, x_teste, y_teste, cv = k)
scores_ada = cross_val_score(ada, x_teste, y_teste, cv = k)
scores_ovr = cross_val_score(ovr, x_teste, y_teste, cv = k)

print("Results:")
print("Decision Tree")
taxa_de_acerto = np.mean(scores_tr)
mean_cntr_tr, var_cntr_tr, std_cntr_tr  = bayes_mvs(scores_tr,0.95)
print("Media de acerto",mean_cntr_tr[0])
print("Desvio padrão:",std_cntr_tr[0])
print("Variância",var_cntr_tr[0])


print("Results:")
print("One Vs Rest")
mean_cntr_ovr, var_cntr_ovr, std_cntr_ovr  = bayes_mvs(scores_ovr,0.95)
print("Media de acerto",mean_cntr_ovr[0])
print("Desvio padrão:",std_cntr_ovr[0])
print("Variância",var_cntr_ovr[0])


print("Naïve Bayes")
mean_cntr_nb, var_cntr_nb, std_cntr_nb  = bayes_mvs(scores_nb,0.95)
print("Media de acerto",mean_cntr_nb[0])
print("Desvio padrão:",var_cntr_nb[0])
print("Variância",std_cntr_nb[0])

print("AdaBoosting classifier")
mean_cntr_ada, var_cntr_ada, std_cntr_ada  = bayes_mvs(scores_ada,0.95)
print("Media de acerto",mean_cntr_ada[0])
print("Desvio padrão:",var_cntr_ada[0])
print("Variância",std_cntr_ada[0])
```

```python
# b. Exiba a matriz de confusão dos resultados;
from sklearn.metrics import confusion_matrix

y_pred_tr = tr.fit(x_treino, y_treino).predict(x_teste)
y_pred_nb = nb.fit(x_treino, y_treino).predict(x_teste)
y_pred_ada = ada.fit(x_treino, y_treino).predict(x_teste)
y_pred_ovr = ovr.fit(x_treino, y_treino).predict(x_teste)

# Compute confusion matrix
print("Decision Tree")
print(pd.crosstab(y_teste, y_pred_tr, rownames=['Real'], colnames=['Pretict'],margins=True))
print("\n")
print("Naïve Bayes")
print(pd.crosstab(y_teste, y_pred_nb, rownames=['Real'], colnames=['Pretict'],margins=True))
print("\n")
print("AdaBoosting")
print(pd.crosstab(y_teste, y_pred_ada, rownames=['Real'], colnames=['Pretict'],margins=True))
print("\n")
print("OneVsRest")
print(pd.crosstab(y_teste, y_pred_ovr, rownames=['Real'], colnames=['Pretict'],margins=True))


```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,zero_one_loss
# c. Avalie as métricas de acurácia, precisão e recall;
print("Decision Tree")
print("Acurácia: ",accuracy_score(y_teste, y_pred_tr))
print("Precisão: ",precision_score(y_teste, y_pred_tr, average='macro'))
print("Recall: ",recall_score(y_teste, y_pred_tr, average='micro') )
print("Naïve Bayes")
print("Acurácia: ",accuracy_score(y_teste, y_pred_nb))
print("Precisão: ",precision_score(y_teste, y_pred_nb, average='macro') )
print("Recall: ",recall_score(y_teste, y_pred_nb, average='micro') )
print("AdaBoosting")
print("Acurácia: ",accuracy_score(y_teste, y_pred_ada))
print("Precisão: ",precision_score(y_teste, y_pred_ada, average='macro') )
print("Recall: ",recall_score(y_teste, y_pred_ada, average='micro') )
print("OneVsRest")
print("Acurácia: ",accuracy_score(y_teste, y_pred_ovr))
print("Precisão: ",precision_score(y_teste, y_pred_ovr, average='macro') )
print("Recall: ",recall_score(y_teste, y_pred_ovr, average='micro') )
```

```python
# d. Analise pelo menos mais duas métricas;

print("Decision Tree")
print("F1: ",f1_score(y_teste, y_pred_tr, average='macro'))
print("Perda de Zero ou Um: ",zero_one_loss(y_teste, y_pred_tr ))
print("Naïve Bayes")

print("F1: ",f1_score(y_teste, y_pred_nb, average='macro'))
print("Perda de Zero ou Um: ",zero_one_loss(y_teste, y_pred_nb ))
print("Logistic Regression")

print("F1: ",f1_score(y_teste, y_pred_ada, average='macro'))
print("Perda de Zero ou Um: ",zero_one_loss(y_teste, y_pred_ada ))

print("F1: ",f1_score(y_teste, y_pred_ovr, average='macro'))
print("Perda de Zero ou Um: ",zero_one_loss(y_teste, y_pred_ovr ))
```
